```python
import pandas as pd

articles = pd.read_csv("Web_Page/Input Data/reference.csv")
```

```python
journals=articles.groupby('First Author',as_index=False).count().sort_values(by='Title',ascending=False)
#journals.sort_values()
journals.head(20)
```

```python
values=[]
labels=[]

for i in range(5):
    value=journals.iloc[i]['Authors']
    label = journals.iloc[i]['Journals']
    
    values.append(value)
    labels.append(label)
    
    i=i+1
    
print(values)
print(labels)
```
